```python
import cv2 as cv
import numpy as np
import os 
import time
from hand_tracking_module import handDetector
```

```python
url='http://192.168.10.8:4747/video'
```

```python
mode =None  #drawing 1 is selection
drawingColor=(0,0,255)
imgCanvas=np.zeros((480,640,3),np.uint8)
```

```python
myList=os.listdir('header')
overlayList=[]
for imPath in myList:
    image=cv.imread(f'header/{imPath}')
    overlayList.append(image)
```

```python
cap=cv.VideoCapture(0)
cap.set(3,1280)
cap.set(4,720)
detector=handDetector(detectionConfidence=0.75)
while True:
    success,img=cap.read()
    img=cv.flip(img,1)
    img=detector.drawHands(img)
    lmList=detector.findPoints(img,draw=False)
    header=overlayList[3]
    if len(lmList) !=0:
        x1,y1=lmList[8] 
        x2,y2=lmList[12]
        print(x1,y1)
        if y1 < lmList[6][1]:
            mode=0
            if y2<lmList[10][1]:
                mode=1
        else: 
            mode=None
    
    if mode !=None:
        if mode==0:
            cv.putText(img,"Draw",(10,120),cv.FONT_HERSHEY_DUPLEX,1,(0,255,0),2)
            if xp==0 and yp==0:
                    xp=x1  
                    yp=y1
            cv.line(img,(xp,yp),(x1,y1),drawingColor,10)
            cv.line(imgCanvas,(xp,yp),(x1,y1),drawingColor,10)
            xp,yp=x1,y1
            
        elif mode==1:
            xp,yp=0,0
            cv.putText(img,"Select",(10,120),cv.FONT_HERSHEY_DUPLEX,1,(0,255,0),2)
            if y1<100:
                if 90<x1<200:
                    header=overlayList[0]
                    drawingColor=(0,0,255)
                elif 255<x1<350:
                    header=overlayList[1]
                    drawingColor=(0,255,0)
                elif 400<x1<495:
                    header=overlayList[2]
                    drawingColor=(0,255,255)
                elif 540<x1<640:
                    header=overlayList[3]
                    drawingColor=(0,0,0)
        cv.circle(img,(x1,y1),10,drawingColor,-1)
    img[0:100,:]=header
    
    imgGray=cv.cvtColor(imgCanvas,cv.COLOR_BGR2GRAY)
    _,INV=cv.threshold(imgGray,50,255,cv.THRESH_BINARY_INV)
    imgINV=cv.cvtColor(INV,cv.COLOR_GRAY2BGR)
    img=cv.bitwise_and(img,imgINV)
    img=cv.bitwise_or(img,imgCanvas)
    cv.imshow('Video',img)
    # cv.imshow('Canvas',imgCanvas)
    # cv.imshow('INV',INV)

    if cv.waitKey(10) & 0xFF==ord('d'):
        break

    
cap.release()
cv.destroyAllWindows()
```
